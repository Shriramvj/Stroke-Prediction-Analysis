```python
#Checking for outliers ✏️
#counting outliers for bmi

bmi_outliers=df[df['bmi']>50]
bmi_outliers['bmi'].shape 
```

```python
#replacing outlier entries with mean of bmi
df["bmi"] = df["bmi"].apply(lambda x: df.bmi.median() if x>50 else x)
```

```python
# Replacing null values of bmi with mean of bmi column

df.bmi.replace(to_replace=np.nan,value=df.bmi.mean(), inplace=True)
print (df.shape)
```

The labels for glucose levels can vary depending on the specific reference range used by healthcare providers and organizations. However, in general, the following labels are often used:

Low Glucose: Less than 70 mg/dL

Normal Glucose: 70 to 99 mg/dL

Medium/Borderline High Glucose: 100 to 125 mg/dL

High Glucose: 126 mg/dL or higher (after two separate tests)

It is important to note that these ranges may vary based on factors such as age, pregnancy, and other health conditions. It is best to consult a healthcare provider to determine an appropriate glucose range for an individual's specific health needs.

```python
# to add a new column to the existing dataset and based on the inputs and ranges of glucose levels encode them into range column as low sugar, medium sugar and high sugar
```

```python
import pandas as pd

# Define the glucose level ranges
low_sugar = (df['avg_glucose_level'] < 70)
medium_sugar = ((df['avg_glucose_level'] >= 70) & (df['avg_glucose_level'] < 140))
high_sugar = (df['avg_glucose_level'] >= 140)

# Add a new column 'sugar_range' to the existing dataset 'df' based on the glucose level ranges
df['sugar_range'] = ''
df.loc[low_sugar, 'sugar_range'] = 'Low Sugar'
df.loc[medium_sugar, 'sugar_range'] = 'Medium Sugar'
df.loc[high_sugar, 'sugar_range'] = 'High Sugar'
```
